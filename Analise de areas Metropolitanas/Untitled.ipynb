```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/Standard_Metropolitan_Areas_Data-data.csv") #carregando o dataset
```

```python
df.head()
```

```python
df.info()
```

```python
plt.scatter(df.crime_rate, df.percent_senior);#criando um análise bivariada entre Taxa de crime e percent senior
plt.show();
```

```python
#É possivel ver que não existe uma relaçao entre os dois o pontos estão bem dispersos, com dois pontos muito foram do comum
#cabe uma analise mais detalhada pra ver o que acontece ali.
```

```python
plt.scatter(df.crime_rate, df.percent_senior);
plt.title('Plot of Crime Rate vs Percent Senior') # Titulo do grafico
plt.xlabel("Crime Rate") # Eixo x
plt.ylabel("Percent Senior") # Eixo y
plt.show()
```

```python
#Podemos mudar o tamanho do grafico/janela usando o figsize

plt.figure(figsize=(12,5)) # 12x5 

#Criando um grafico linear
plt.plot(df['work_force'], df['income']) 
plt.xlabel("Work Force")
plt.ylabel("Income")
plt.show()
```

```python
#Tambem podemos usaro subplot para plotar mais de um grafico

plt.subplot(1,2,1)  
plt.plot(df.work_force, df.income, "go")
plt.title("Income vs Work Force")

plt.subplot(1,2,2).label_outer() #Aqui podemos separar em dois graficos diferente ou plotar apenas um gráfico a fim de comparacao

plt.plot(df.hospital_beds, df.income, "r^")
plt.title("Income vs Hospital Beds")

plt.suptitle("Sub Plots") 
plt.show()
```

```python
#Plotando um histograma
plt.title("Histograma")
plt.xlabel("Porcentagem de  Senior Citizens")
plt.ylabel("Frequencia")

plt.hist(df.percent_senior)
plt.show()
```

```python
#Por ultimo um grafico de barras de CRIMES POR REGIAO

plt.bar(df.region, df.crime_rate, color="red")

plt.title("Grafico de barras")
plt.xlabel("Regiao")
plt.ylabel("Taxa de crime")
plt.show()
```
