# Uma breve analise sobre os profissionais da IBM, procurando entender as relacoes da empresa, e respondendo perguntas

Comecamos importando os modulos que iremos usar.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rh = pd.read_csv('rh_ibm.xls')
```

```python
rh.info()
```

# Análise Unívariada 

```python
plt.hist(rh.Age);
plt.title('Histograma de Idade');
plt.ylabel('Numero de profissionais');
plt.xlabel('Idade');
```

```python
plt.hist(rh.MonthlyIncome);
plt.title('Histograma de Salario');
```

```python
rh.Gender.value_counts()
#Numero de profissionais por genero
```

```python
rh.Gender.value_counts(1)
#Numero de profissionais por genero em porcentagem
```

```python
rh.EducationField.value_counts(1).plot.bar();
plt.ylabel('Numero de funcionarios');
```

# Análise Bivariada ou Multivariada

```python
sns.boxplot(x='Gender',y='MonthlyIncome',data = rh);
```

```python
# É possível notar que existe um equilibrio de salario na IBM, o que é bom , é mais comum vermos discrepância
```

```python
#Pegando todas as colunas do tipo númerica para plotar uma correlacao
colunas_num = []
for column in rh.columns:
    if rh[column].dtype != 'object':
       colunas_num.append(column)
```

```python
rh[colunas_num].corr()
```
